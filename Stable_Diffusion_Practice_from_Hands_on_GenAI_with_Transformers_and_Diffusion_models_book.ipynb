```python
from genaibook.core import get_device
import torch
from diffusers import StableDiffusionPipeline
```

```python
#Check which device your using to send the pretrained models to
#device = get_device()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
#Load the pretrained model stable-diffusion-v1-5 to generate images
pipe = StableDiffusionPipeline.from_pretrained(
    "stable-diffusion-v1-5/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    variant="fp16",
).to(device)
```

```python
#Create a prompt and send it to the model in order to generate a new image
prompt = "a photograph of an astronaut riding a horse"
pipe(prompt).images[0]
```
